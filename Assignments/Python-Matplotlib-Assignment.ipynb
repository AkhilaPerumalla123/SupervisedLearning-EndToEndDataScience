```python
# 10. In the diabetes dataset, plot a pie-chart showing % of diabetes and % of non-diabetes.

diabetes.head()
```

```python
fig = plt.figure(figsize=(10,10))
labels='Diabetes','No-diabetes'
diabetes_count = diabetes[diabetes["Outcome"]==1]["Outcome"].count()
non_diabetes_count = diabetes[diabetes["Outcome"]==0]["Outcome"].count()
my_explode = (0.1, 0)
plt.pie([diabetes_count,non_diabetes_count],labels=labels,autopct='%1.1f%%',explode=my_explode)
plt.legend(loc ="upper left")
plt.title("Diabetes Vs Non-diabetes",size=20)
# diabetes["Outcome"].value_counts().plot.pie()
plt.show()
```
